````python
from openai import OpenAI
import os

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY")) #não é mais 'openai.api_key = openai_api_key' pois a API mudou

def generate_ai_news(user):
    response = client.chat.completions.create( #nao é mais 'openai.ChatCompletion.create()' pois a API mudou (openai>=1.0)
        model="gpt-4o-mini",
        messages=[
            {
                "role": "system",
                "content": "Você é a Fernanda, especialista em marketing bancário. Tom amigável e direto."
            },
            {
                "role": "user",
                "content": f"Crie uma mensagem para {user['nome']} sobre a importância dos investimentos (máximo de 80 caracteres)"
            }
        ]
    )

    return response.choices[0].message.content.strip('"')


for user in users:
    user['news'] = [] #limpa antes pra nao mandar mensagens repetidamente para a mesma pessoa, por conta da quantidade de execuções
    news = generate_ai_news(user)

    print(news)
    user['news'].append({
        "icon": "https://raw.githubusercontent.com/brandaonanda/santander-dev-week-2023-api/refs/heads/main/docs/icons/credit.svg",
        "description": news
    })
````

````python
import pandas as pd

#carrega o CSV original
df = pd.read_csv(url)
users = df.to_dict(orient="records")

#processamento
for user in users:
    try:
        news = generate_ai_news(user)

        user['news'] = news  # salva só o texto (melhor pra CSV)

        success = True

    except Exception as e:
        print(f"Erro com {user['nome']}: {e}")
        user['news'] = None
        success = False

    print(f"User {user['nome']} processed? {success}!")


df_final = pd.DataFrame(users) #cria novo DataFrame atualizado

#salva novo CSV
df_final.to_csv("dadosSDW2023NEW_atualizado.csv", index=False)

print("\nCSV atualizado salvo com sucesso!")
````
